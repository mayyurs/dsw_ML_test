# Performing EDA on dataset

```python
# import all necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the historic data
historic_data = pd.read_csv('historic.csv')
```

```python
# Display the first few rows of the dataset
print("First few rows of historic data:")
print(historic_data.head())
```

The data set has 6 major columns/features : item_no,category,     main_promotion, color, stars and success_indicator.

```python
# Types of data
historic_data.dtypes
```

```python
# Summary statistics
print("Summary statistics:")
print(historic_data.describe())
```

```python
# Distribution of success_indicator (target variable)
plt.figure(figsize=(8, 6))
sns.countplot(x='success_indicator', data=historic_data)
plt.title('Distribution of Success Indicator')
plt.xlabel('Success Indicator')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of categories
plt.figure(figsize=(10, 6))
sns.countplot(x='category', data=historic_data)
plt.title('Distribution of Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of main_promotion
plt.figure(figsize=(8, 6))
sns.countplot(x='main_promotion', data=historic_data)
plt.title('Distribution of Main Promotion')
plt.xlabel('Main Promotion')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of colors
plt.figure(figsize=(10, 6))
sns.countplot(x='color', data=historic_data)
plt.title('Distribution of Colors')
plt.xlabel('Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of stars
plt.figure(figsize=(8, 6))
sns.histplot(x='stars', data=historic_data, bins=20, kde=True)
plt.title('Distribution of Stars')
plt.xlabel('Stars')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot of 'stars' by 'success_indicator'
plt.figure(figsize=(10, 6))
sns.boxplot(data=historic_data, x='success_indicator', y='stars')
plt.title('Stars by Success Indicator')
plt.xlabel('Success Indicator')
plt.ylabel('Stars')
plt.show()
```

```python
# Box plot of 'price' by 'success_indicator'
plt.figure(figsize=(10, 6))
sns.boxplot(data=historic_data, x='success_indicator', y='color')
plt.title('Colors by Success Indicator')
plt.xlabel('Success Indicator')
plt.ylabel('Colors')
plt.show()
```

```python
# Box plot of 'stars' by 'category'
plt.figure(figsize=(12, 6))
sns.boxplot(data=historic_data, x='category', y='stars')
plt.title('Stars by Category')
plt.xlabel('Category')
plt.ylabel('Stars')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram of 'stars'
plt.figure(figsize=(10, 6))
plt.hist(historic_data['stars'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Stars')
plt.xlabel('Stars')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Histogram of 'category'
plt.figure(figsize=(10, 6))
plt.hist(historic_data['category'], bins=len(historic_data['category'].unique()), color='lightgreen', edgecolor='black', alpha=0.7)
plt.title('Histogram of Category')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Bar plot of 'main_promotion'
plt.figure(figsize=(10, 6))
sns.countplot(data=historic_data, x='main_promotion', palette='pastel')
plt.title('Bar Plot of Main Promotion')
plt.xlabel('Main Promotion')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot of 'color'
plt.figure(figsize=(10, 6))
sns.countplot(data=historic_data, x='color', palette='pastel')
plt.title('Bar Plot of Color')
plt.xlabel('Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pairplot for numerical variables
sns.pairplot(historic_data[['category', 'main_promotion', 'color', 'stars', 'success_indicator']], hue='success_indicator')
plt.title('Pairplot of Numerical Variables')
plt.show()
```

```python
# Correlation Analysis
plt.figure(figsize=(10, 8))
sns.heatmap(historic_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Distribution of Stars by Success Indicator
plt.figure(figsize=(10, 6))
sns.histplot(data=historic_data, x='stars', hue='success_indicator', kde=True, bins=20)
plt.title('Distribution of Stars by Success Indicator')
plt.xlabel('Stars')
plt.ylabel('Frequency')
plt.legend(title='Success Indicator')
plt.show()
```

```python
# Success by Category
plt.figure(figsize=(10, 6))
sns.countplot(data=historic_data, x='category', hue='success_indicator')
plt.title('Success by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Success Indicator')
plt.show()
```

```python
# Success by Main Promotion
plt.figure(figsize=(10, 6))
sns.countplot(data=historic_data, x='main_promotion', hue='success_indicator')
plt.title('Success by Main Promotion')
plt.xlabel('Main Promotion')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Success Indicator')
plt.show()
```

```python
# Success by Color
plt.figure(figsize=(10, 6))
sns.countplot(data=historic_data, x='color', hue='success_indicator')
plt.title('Success by Color')
plt.xlabel('Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Success Indicator')
plt.show()
```

```python
# Pairwise Relationships
sns.pairplot(historic_data[['stars', 'category', 'main_promotion', 'color', 'success_indicator']], hue='success_indicator')
plt.suptitle('Pairwise Relationships')
plt.show()
```

```python
# Outlier Detection
# For example, for numerical features like 'stars':
plt.figure(figsize=(10, 6))
sns.boxplot(data=historic_data, x='success_indicator', y='stars')
plt.title('Outlier Detection')
plt.xlabel('Success Indicator')
plt.ylabel('Stars')
plt.show()
```
